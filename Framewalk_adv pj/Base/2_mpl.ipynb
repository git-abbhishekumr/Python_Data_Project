```python
#Importing Libraries
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
datasets = load_dataset('lukebarousse/data_jobs')
df=datasets['train'].to_pandas()


#Data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
date_count=df.job_posted_date.value_counts()
plt.plot(date_count.index,date_count)
```

```python
date_count= date_count.sort_index()
plt.plot(date_count.index,date_count)
```

```python
df['Job_posted_month']=df['job_posted_date'].dt.month
df
```

```python
monthly_count = df.Job_posted_month.value_counts().sort_index()
monthly_count
```

```python
plt.plot(monthly_count.index,monthly_count.val)  
```

```python
job_count=df.job_title_short.value_counts()
job_count
df
```

```python
plt.bar(job_count.index, job_count)
plt.title('Posting by Job Title')
plt.ylabel('Job Count')
plt.xlabel('Job Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Ensure labels fit within the plot area
plt.show()  # Display the plot
```
